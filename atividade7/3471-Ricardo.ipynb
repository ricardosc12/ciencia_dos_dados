```python
#### Não altere essa célula!!! ####
import pandas as pd
import string
import random
import numpy as np

from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

from learntools_dados_ufv.core import binder; binder.bind(globals())
from learntools_dados_ufv.ccf425.pratica_7b import *
```

# Roteiro

## Estruturando os dados

### Criar o conjunto de dados

```python
dataset = [['Leite', 'Cebola', 'Nutela', 'Feijão', 'Ovo', 'Iogurte'],
          ['Arroz', 'Cebola', 'Nutela', 'Feijão', 'Ovo', 'Iogurte'],
          ['Leite', 'Maça', 'Feijão', 'Ovo'],
          ['Leite', 'Chocolate', 'Pipoca', 'Feijão', 'Iogurte'],
          ['Pipoca', 'Cebola', 'Feijão', 'Sorvete', 'Ovo']]

dataset
```

### Transformar no formato para ser utilizado no Apriori

```python
transactions_oht = TransactionEncoder()
transactions = transactions_oht.fit(dataset).transform(dataset)
transactions
```

```python
df = pd.DataFrame(transactions, columns=transactions_oht.columns_)
df
```

## Calcular conjunto de itens frequentes

Com suporte mínimo de 0.6

```python
# Calcula os conjuntos de itens frequentes, com suporte minimo de 0.6
frequent_itemsets = apriori(df, min_support=0.6, use_colnames=True)
frequent_itemsets
```

## Calcular regras de associacão utilizando o Apriori

Com filtro de confiança de 0.1

```python
rules = association_rules(frequent_itemsets, metric='confidence', min_threshold=0.1)
rules
```

```python
# Filtro por 'confidence'. Pode ser feito por outros campos também
rules[rules['confidence'] > 0.75]
```

- **antecedent support** *(0.033729)*: Frequência relativa com que a combinação(itemset) {"pancakes", "mineral water"} aparecem nas transações, 3.37%.

- **support** *(0.17411)*: Frequência relativa com que a combinação(itemset) {"pancakes", "mineral water", "spaghetti"} aparecem nas transações, ou seja, freq(XUY)/|T|, 17.41%.

- **lift** *(1.952333)*: O lift demonstra o quanto a presença do itemset {"pancakes", "mineral water"} na transação aumenta a probabilidade de conter {"spaghetti"}. Um valor maior que 1 quer dizer que o cliente comprando panquecas e água aumentam a probabilidade dele também levar espaguete.

### 4.c. Utilize a função auxiliar abaixo para calcular conjuntos de itens frequentes (utilize suporte mínimo no intervalo [0.05, 0.3]) variando os tamanhos. Qual padrão você observa?
Qual variável você acredita que afeta o desempenho desse método? Experimente e discorra sobre quando usar esse método.

```python
# Gera uma matriz 2D de strings aleatórias.
#
# Parâmetro dv (different values): Quantidade de strings aleatórias únicas que a matriz irá conter.
# Parâmetro vpi (values per instance): Quantidade de strings aleatórias que cada instância terá.
# Parâmetro n: Quantidade de instâncias que será retornada.
def generate_fictional_data(dv=50, vpi=10, n=2500000):
    instances = [''.join(random.choices(string.ascii_uppercase + string.digits, k=20)) for _i in range(dv)]
    fictional_data = [random.choices(instances, k=vpi) for _i in range(n)]
    return fictional_data
```

```python
def generate_frequent_itemsets(data):
  transactions = transactions_oht.fit(data).transform(data)
  df_data = pd.DataFrame(transactions, columns=transactions_oht.columns_)
  return apriori(df_data, min_support=0.05, use_colnames=True)
```

```python

frequent_itemsets_2500000 = generate_frequent_itemsets(generate_fictional_data(n=2500000))
frequent_itemsets_250000 = generate_frequent_itemsets(generate_fictional_data(n=250000))
frequent_itemsets_25000 = generate_frequent_itemsets(generate_fictional_data(n=25000))
frequent_itemsets_2500 = generate_frequent_itemsets(generate_fictional_data(n=2500))
frequent_itemsets_250 = generate_frequent_itemsets(generate_fictional_data(n=250))
frequent_itemsets_25 = generate_frequent_itemsets(generate_fictional_data(n=25))

print("2500000 Strings")
print(frequent_itemsets_2500000[frequent_itemsets_2500000['support']<=0.3]['support'].describe())
print("250000 Strings")
print(frequent_itemsets_250000[frequent_itemsets_250000['support']<=0.3]['support'].describe())
print("25000 Strings")
print(frequent_itemsets_25000[frequent_itemsets_25000['support']<=0.3]['support'].describe())
print("2500 Strings")
print(frequent_itemsets_2500[frequent_itemsets_2500['support']<=0.3]['support'].describe())
print("250 Strings")
print(frequent_itemsets_250[frequent_itemsets_250['support']<=0.3]['support'].describe())
print("25 Strings")
print(frequent_itemsets_25[frequent_itemsets_25['support']<=0.3]['support'].describe())
```

As informações acima descrevem o comportamento do suporte conforme ajustamos o tamanho do array de strings aleatórias.

A geração de strings aleatórias tem um limite para o número de strings aleatórias únicas. Isso é observado quando elevamos o número de combinações aleatórias, as keys geradas têm suporte alto, o que significa que aparecem constantemente em diferentes "transações". O parametro dv, se ajustado deverá
diminuir esse suporte novamente para os valores altos, entretanto, demorará para ser executado

Com uma média com baixo desvio padrão, um número alto gera keys que aparecem 18% nos dados. Ou seja, baixo caráter aleatório de keys.
